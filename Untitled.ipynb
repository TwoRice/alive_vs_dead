### Imports

```python
import pandas as pd
from decimal import Decimal
```

### Constants

```python
POPULATION_PATH = "population.csv"
LIFE_EXP_PATH = "life_expectancy.csv"
EVER_LIVED = 117e09
```

### Functions

### Load Data

```python
population = pd.read_csv(POPULATION_PATH)
population = population[population.Entity == "World"]
population.rename({"Population (historical estimates)": "population", "Year": "year"}, axis=1, inplace=True)
population = population[["year", "population"]].reset_index(drop=True)
print(population.shape)
population
```

```python
life_exp = pd.read_csv(LIFE_EXP_PATH)
life_exp = life_exp[life_exp.Entity == "World"]
life_exp.rename({"Life expectancy at birth (historical)": "life_exp", "Year": "year"}, axis=1, inplace=True)
life_exp = life_exp[["year", "life_exp"]].reset_index(drop=True)
life_exp
```

### Convert to Yearly Data

#### Population

```python
population["time_interval"] = population.year.diff(periods=-1) * -1
population["growth"] = population.population.diff(periods=-1) * -1
population["yearly_growth"] = population.apply(lambda x: Decimal(x.growth / x.time_interval), axis=1)
population.fillna(0, inplace=True)
population
```

```python
new_years = range(-200000, 2022)
new_population = []
yearly_growth = 0
current_population = 0
for year in new_years:
    if year in population.year.values:
        current_population = population[population.year == year].population.values[0]
        yearly_growth = population[population.year == year].yearly_growth.values[0]    
    else:
        current_population = round(current_population + Decimal(yearly_growth))

    new_population.append(current_population)
                              

yearly_population = pd.DataFrame({"year": new_years, "population": new_population})
yearly_population["growth"] = yearly_population.population.diff()
yearly_population
```

```python
yearly_population.plot(x="year", y="growth")
```

#### Life Expectancy

```python
new_years = range(-200000, 2022)
new_life_exp = []
current_life_exp = 25
for year in new_years:
    if year in life_exp.year.values:
        current_life_exp = life_exp[life_exp.year == year].life_exp.values[0]
        
    new_life_exp.append(round(current_life_exp))
    
yearly_life_exp = pd.DataFrame({"year": new_years, "life_exp": new_life_exp})
yearly_life_exp
```

```python
yearly_life_exp.plot(x="year", y="life_exp")
```

### Merge

```python
yearly_stats = pd.merge(yearly_population, yearly_life_exp)
yearly_stats
```

### Do the Thing?
For now we will assume everyone in 10,000BCE was 0 years old. Obviously this is a bad assumption, I might revisit this later, but probably not.

```python
yearly_deaths = [0]
first_year = yearly_stats.loc[0]
death_timers = [(first_year.life_exp, first_year.population)]
for _, year in list(yearly_stats.iterrows())[1:]:
    # decrement death timers
    new_death_timers = []
    died_this_year = 0
    for timer in death_timers:
        if timer[0] - 1 < 1:
            died_this_year += timer[1]
        else:
            new_death_timers.append((timer[0] - 1, timer[1]))
    yearly_deaths.append(died_this_year)
    death_timers = new_death_timers

    death_timers.append((year.life_exp, year.growth))
```

```python
yearly_stats["deaths"] = yearly_deaths
yearly_stats["total_dead"] = yearly_stats.deaths.cumsum()
```
